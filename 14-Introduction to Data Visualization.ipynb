# Introduction to Data Visualization

## Introduction to Popular Data Visualization Libraries
1. Matplotlib

```python
# import library dengan alias plt
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Membuat scatter plot
plt.scatter(x, y)

# Menampilkan plot
plt.show()
```

2. Seaborn

```python
# import library seaborn dengan alias sns
import seaborn as sns

# Data
data = [1, 1, 2, 2, 2, 3, 3, 4, 4, 5]

# Membuat histogram
sns.histplot(data)

# Menampilkan plot
plt.show()
```
