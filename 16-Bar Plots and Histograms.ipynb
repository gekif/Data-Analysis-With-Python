# Bar Plots and Histograms

## Creating bar plots and customization options

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Membuat bar plot
plt.bar(categories, values, color='blue')

# Memberikan judul dan label sumbu
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Menampilkan plot
plt.show()
```

## Histograms and Density Plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = np.random.randn(1000)

# Membuat histogram
plt.hist(data, bins=30, color='green')

# Memberikan judul dan label sumbu
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Menampilkan plot
plt.show()
```
