# Pie Charts and Box Plots

## Creating Pie Charts and Customization Options

```python
import matplotlib.pyplot as plt

# Data
labels = ['A', 'B', 'C', 'D']
sizes = [30, 25, 20, 25]

# Membuat pie chart
plt.pie(sizes, labels=labels, colors=['red', 'blue', 'green', 'orange'], autopct='%1.1f%%')

# Memberikan judul
plt.title('Pie Chart')

# Menampilkan plot
plt.show()
```

## Box Plots and Whisker Plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Membuat box plot
plt.boxplot(data, tick_labels=['A', 'B', 'C'])

# Memberikan judul dan label sumbu
plt.title('Box Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Menampilkan plot
plt.show()
```
