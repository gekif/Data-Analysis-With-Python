# Line Plots and Scatter Plots

## Creating Line Plots using Matplotlib

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Membuat line plot
plt.plot(x, y)

# Memberikan judul dan label sumbu
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Menampilkan plot
plt.show()
```

## Scatter Plots and Customization Options

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Membuat scatter plot
plt.scatter(x, y, color='red', marker='o')

# Memberikan judul dan label sumbu
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Menampilkan plot
plt.show()
```
