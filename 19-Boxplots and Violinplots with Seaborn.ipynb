# Boxplots and Violinplots with Seaborn

## Creating boxplots

```python
import pandas as pd

# Membuat data frame
data = pd.DataFrame({
    'Nama': ['Andi', 'Budi', 'Caca', 'Deni', 'Eva'],
    'Usia': [25, 30, 22, 28, 24],
    'Pendapatan': [5000, 6000, 4500, 5500, 5200],
    'Jenis_Pekerjaan': ["Developer", "Developer", "Developer", "Designer", "Designer"]
})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Jenis_Pekerjaan', y='Usia', data=data)
plt.title('Perbandingan Usia antara Jenis Pekerjaan')
plt.xlabel('Jenis Pekerjaan')
plt.ylabel('Usia')
plt.show()
```

## Creating Violinplots

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.violinplot(x='Jenis_Pekerjaan', y='Pendapatan', data=data)
plt.title('Perbandingan Pendapatan antara Jenis Pekerjaan')
plt.xlabel('Jenis Pekerjaan')
plt.ylabel('Pendapatan')
plt.show()
```
