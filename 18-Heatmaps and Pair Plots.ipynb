# Heatmaps and Pair Plots

## Creating Heatmaps

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = np.random.rand(5, 5)

# Membuat heatmap
plt.imshow(data, cmap='hot')

# Menambahkan colorbar
plt.colorbar()

# Memberikan judul
plt.title('Heatmap')

# Menampilkan plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Data
data = pd.DataFrame(data)

# Membuat pair plot
sns.pairplot(data)

# Menampilkan plot
plt.show()
```
